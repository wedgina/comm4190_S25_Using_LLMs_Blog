#### Critiques

I find the way the AI responds to be very interesting, particularly its accent. When speaking about non english based languages and AI in ChatGPT we questioned where it was getting its information from, especially for a language that's not really represented online. I ultimately decided, that google translate has been able to translate to and from Haitian Creole for years and it is most likely getting its data from there. As a person who grew up speaking it my entire life, I found the AI's accent to be interersting. It sounded incredibly frenchy to me, in which, that's not impossible. Haiti is the most populated Caribbean island and there is various regional differences between native speakers, and especially in the north, where people tend to sound a lot more frenchy. But I don't know this can also sound different to me, because it sounds very much like a robot talking as well. It uses very little slang to communicate with me, when I used a good amount of slang to communicate with it. When I speak Creole I use a lot of slang and merge words which is natural for me and many, but when it began the convo, it was incredibly formal and used the long way to get out its words.

**Example**

* It originally says "wi mwen kapab pale Kreyol" but when I speak and it realizes how I speak it changes it to "wi mwen ka pale kreyol"

Its speaking ability was actually decent. It spoke Creole pretty clearly and in a way that made sense for most of the conversation. However, its comprehension was very inadequate. While with speaking it understood me enough to keep interacting with me its transcription of its comprehension was pretty bad. Below I have examples that showcase where it messed up pretty badly. 

**Examples**

> 1. Okonemi Fekfi Ngadon Shokele, Manifest, Mwen Mwenopiu

This sentence occured when I was talking about shows I watched. I know when this sentenced happened, and what I truly said but I only understand like a quarter of this sentence. It doesn't transcribe what I am saying to Kreyol accurately. This sentence doesn't even look like Kreyol or even latin based, which Kreyol is. 

> Ou konnen mwen fek fin gade yon show ki rele Manifest? Mwen renmen anpil.

I am actually saying this: You know, I just finished watching a show called Manifest? I like it a lot. 

> 2. Merci d'avoir regarde cette video, n'hesitez pas a la partager avec vos amis et a vous abonner a la chaine pour ne manquer aucune de mes nouvelles videos

Here, the sentence appears to be mostly written in French, which is closer than what the previous example showed but yo it's not even close to what I was saying. This sentence says "Thank you for watching this video, don't hesitate to share it with your friends and subscribe to the channel so you don't miss any of my new videos". At NO point in this conversation I ever said this. This is what the sentence is supposed to truly look like had it had been transcribed correctly:

> Ou konnen mwen te byen satisfè avek kijan li te fini. Gen anpil moun ki di yo pa renmen kijan li te fini, men mwen te renmen anpil

Which means, "You know I was very satisfied with how it ended. A lot of people said they did not like it but I actually did. I liked it a lot"

There is very much a clear difference in the alphabetic/grammar structures that compose the two. I'm curious as to why the AI when it was speaking was able to understand and converse back to me in Kreyol despite the written transcription saying that I was speaking French and saying that I'm saying something completely different. Like why are both inconsistent with each other? Perhaps training data from google translate or something is easier than the AI recognizing Creole being spoken. I kind of doubt it's been trained on how Creole sounds well enough to transcribe it accurately. I think this broadly reflects its societal implications. Many do not see Creole as its own language and mistakenly so see it as a dying or irrelevant language. This is one of those things that reflect the people behind these systems, as I seriously doubt they really had representation behind the scenes or the care to be educated about how Creole is its own language apart from French. 

> 3. Pa gen kisa!

This phrase actually makes no sense. I guess it's attempting to use slang here to say "no problem", but instead here it says "there's no what", which makes absolutely no sense. I've never heard this phrase be uttered in a normal setting. It is better for it to have said "pa gen pwoblèm"
