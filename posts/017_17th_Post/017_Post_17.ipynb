# Can AI Generate as Strong of a Memory and Experience as we can?

```python
from dotenv import load_dotenv
from anthropic import Anthropic
import base64
import httpx
from IPython.display import display, Image, Markdown

_ = load_dotenv()
client = Anthropic()
```

Flashbulb memories are memories people have regarding how they found out about an event, particularly a shocking one. For those who were alive then, many remember where they were for 9/11 and exactly how they found out about it. These memories tend to evoke quite strong memories and emotions. Due to the passion and strength behind recalling these emotions, many often feel that this is a memory they are able to recall to near perfect. This is not necessarily the case, and actually quite the opposite many times. There is no guarantee to accuracy but it is still interesting how much people are able to collectively recall a memory about an event so vividly. In my blog, I will be using examples of common flashbulb memories and asking Claude AI to generate a possible and believable flashbulb memory as well.

# Memory 1 : 9/11

```python
def generate_memory(prompt, model='claude-3-5-haiku-20241022', max_tokens=1000, temperature=0.2):
    message = client.messages.create(
        model=model,
        max_tokens=max_tokens,
        temperature=temperature,
        messages=[
            {"role": "user", "content": prompt}
        ]
    )
    return message.content[0].text
prompt = "Generate a memory as if someone was remembering exactly where they were, how they found out, and what they were feeling when 9/11 happened"
response_911 = generate_memory(prompt)
print(response_911)
```
