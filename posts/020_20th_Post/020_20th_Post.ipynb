# Writing Creative Stories with Google Gemini

Not sure who remembers what madlibs are, but when I was younger my teachers used to have us doing them for funsies. They're a story filled with a few blanks throughout the passage, of a story that's supposed to be funny and random and you're supposed to fill them in to have a funny story. We used to give random words based on what that blank required (adjective, verbs, noun, etc)---Wait a minute, I just realized that they did this to get us to also practice grammar at the same time. Since LLMs are not humans do they even get humor and creativity like we do? Or will they just choose random nonsense to fill in the blanks? Today I am going to be using Google Gemini to see how it will go about filling in those blanks. 

```python
import google.generativeai as genai
from IPython.display import Image, display, Markdown
from dotenv import load_dotenv
from PIL import Image
import os
_ = load_dotenv()

genai.configure()

model = genai.GenerativeModel(model_name="models/gemini-1.5-pro-latest")

response = model.generate_content(["Can you provide an explanation and evidence to support the question on if the removal of Affirmative Action and DEI is right and determine which side is stronger"])

def create_madlibs(image_path, model_name="models/gemini-1.5-pro-latest"):
    model = genai.GenerativeModel(model_name=model_name)
    image = Image.open(image_path)
    prompt = (
        "This is a Mad Libs story image with blanks. The goal is to have a completed funny and creative story, please return the full completed story."
    )
    response = model.generate_content(
        [prompt, image],
        generation_config=genai.types.GenerationConfig(temperature=0.9)
    )
    return response.text
```

### Madlib 1: A story about zombies




```python
zombies = create_madlibs("zombpie-picnic-1.jpg")

display(Markdown(zombies))
```

### Madlib 2: If you give a ________

```python
ifyougivea = create_madlibs("fun mad lib.jpg") 

display(Markdown(ifyougivea)) 
```

### Madlib 3: A story about space

```python
spacelibs = create_madlibs("madlibs.png")

display(Markdown(spacelibs))
```
